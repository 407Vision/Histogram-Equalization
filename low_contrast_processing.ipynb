```python
# Importing the neccessary libraries
import cv2
import matplotlib.pyplot as plt
import os
```

```python
# Loading the image 
image_path = './low_contrast_image.jpg'

# Checking the existence of the image
print(os.path.exists(image_path))

image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Check image dimensions
print("Image Shape:", image.shape) # (Height, Width, Channels)
print("Image size:", image.size) # Total number of pixels
print("Image data type:", image.dtype) # Example output: uint8 (indicating that the pixel values are 8-bit unsigned integers)
```

```python
print(image[300, 400])
```

```python
# Applying Histogram equalization
hist_eq = cv2.equalizeHist(image)
```

```python
# Applying CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_result = clahe.apply(image)
```

```python
#Visualizing the results
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(2, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('on')

# Histogram Equalization
plt.subplot(2, 3, 2)
plt.title("Histogram Equalization")
plt.imshow(hist_eq, cmap='gray')
plt.axis('on')

# CLAHE
plt.subplot(2, 3, 3)
plt.title("CLAHE")
plt.imshow(clahe_result, cmap='gray')
plt.axis('on')

# Displaying the Histogram graph for real Image
plt.subplot(2, 3, 4)
plt.title("Original Histogram")
plt.hist(image.ravel(), bins=256, range=(0, 256), color='blue')
plt.xlabel("Pixel Values")
plt.ylabel("Frequency")

# Displaying the Histogram graph for Histogram Equalization
plt.subplot(2, 3, 5) 
plt.title("Histogram Equalized")
plt.hist(hist_eq.ravel(), bins=256, range=(0, 256), color='green')
plt.xlabel("Pixel Values")
plt.ylabel("Frequency")

# Displaying the Histogram graph for CLAHE image
plt.subplot(2, 3, 6); plt.title("CLAHE Histogram") 
plt.hist(clahe_result.ravel(), bins=256, range=(0, 256), color='red')
plt.xlabel("Pixel Values")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
